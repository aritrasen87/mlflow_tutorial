```python
import os 
from dotenv import load_dotenv
from langchain.agents import create_agent
from langchain.tools import tool
from langchain_tavily import TavilySearch
from langchain.chat_models import init_chat_model

load_dotenv()  # Load environment variables from .env file

# Now you can access your environment variables using os.environ
os.environ['OPENAI_API_KEY'] = os.environ.get("OPENAI_API_KEY")
os.environ['TAVILY_API_KEY'] = os.environ.get("TAVILY_API_KEY")
```

```python
import mlflow

# Calling autolog for LangChain will enable trace logging.
mlflow.langchain.autolog()

# Optional: Set a tracking URI and an experiment
mlflow.set_experiment("langchain_experiments")
mlflow.set_tracking_uri("http://localhost:5000")
```

```python
llm = init_chat_model("gpt-5-mini-2025-08-07")
```

```python
tavily_tool = TavilySearch(max_results=3)

@tool
def web_search_agent(query: str) -> str:
    """Tool that performs a web search for a given query."""
    results = tavily_tool.run(query)
    return results
```

```python
agent = create_agent(model=llm,
    tools=[web_search_agent],
    system_prompt='You are a helpful assistant that can provide ' \
    'web search results based on user queries.'
)
```

```python
response = agent.invoke({'messages': [{'role':'user',
                                       'content':'When and where is the 2026 FIFA World Cup?'}]})

print(response["messages"][-1].content)
```
